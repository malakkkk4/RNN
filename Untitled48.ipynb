```python
from tensorflow.keras.preprocessing.text import one_hot
import warnings
warnings.filterwarnings('ignore')
```

```python
sent=[ 'the glass of milk',
        'the glass of juice',
        'the cup of tea ',
        'I am a good boy',
       'I am a good developer',
       'understand the meaning of words',
       'your videos are good']
```

```python
voc_size = 10000
```

```python
one_hot_repr =[one_hot(words,voc_size) for words in sent]
one_hot_repr
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs).shape
```

```python
embeddings=model.predict(embedded_docs)
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Reshape embeddings from (batch_size, sequence_length, embedding_dimension) to (batch_size * sequence_length, embedding_dimension)
reshaped_embeddings = embeddings.reshape(-1, embeddings.shape[2])

pca=PCA(n_components=3)
reduced_embeddings =pca.fit_transform(reshaped_embeddings)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Extract the 3 components
x = reduced_embeddings[:, 0]
y = reduced_embeddings[:, 1]
z = reduced_embeddings[:, 2]

# Plot the scatter points
ax.scatter(x, y, z, c='blue', marker='o')

# Set labels and title
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3D PCA of Word Embeddings')

plt.show()
```
